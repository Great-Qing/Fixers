```python
## CNH = 역외 위안화 환율
CNH = pd.read_csv('C:/Users/nueva/Desktop/p1/test/USD_CNH 과거 데이터.csv')

# 종가, 시가 분리 및 평균
CNH_Ed = CNH.iloc[:, 1]
CNH_St = CNH.iloc[:, 2]
avg_CNH = (CNH_Ed + CNH_St) /2 
avg_CNH = np.array(avg_CNH).reshape(-1,1)

# Fit_transform
scaler = StandardScaler()
_avg_CNH = scaler.fit_transform(avg_CNH)
```

```python
## GC = 금 선물
GC = pd.read_csv('C:/Users/nueva/Desktop/p1/test/금 선물 과거 데이터.csv')

# 종가, 시가 분리 및 평균
GC_Ed = GC.iloc[:, 1]
GC_St = GC.iloc[:, 2]
avg_GC = (GC_Ed + GC_St) /2 
avg_GC = np.array(avg_GC).reshape(-1,1)

# Fit_transform
scaler = StandardScaler()
_avg_GC = scaler.fit_transform(avg_GC)
```

```python
## SI = 은 선물
SI = pd.read_csv('C:/Users/nueva/Desktop/p1/test/은 선물 과거 데이터.csv')

# 종가, 시가 분리 및 평균
SI_Ed = SI.iloc[:, 1]
SI_St = SI.iloc[:, 2]
avg_SI = (SI_Ed + SI_St) /2 
avg_SI = np.array(avg_SI).reshape(-1,1)

# Fit_transform
scaler = StandardScaler()
_avg_SI = scaler.fit_transform(avg_SI)
```

```python
## HG  = 구리 선물  
HG= pd.read_csv('C:/Users/nueva/Desktop/p1/test/구리 선물 과거 데이터.csv')

# 종가, 시가 분리 및 평균
HG_Ed = HG.iloc[:, 1]
HG_St = HG.iloc[:, 2]
avg_HG = (HG_Ed + HG_St) /2 
avg_HG = np.array(avg_HG).reshape(-1,1)

# Fit_transform
scaler = StandardScaler()
_avg_HG = scaler.fit_transform(avg_HG)
```

```python
# Display the distance
dist_GC = dtw.distance(_avg_CNH,_avg_GC)
dist_SI = dtw.distance(_avg_CNH,_avg_SI)
dist_WTI = dtw.distance(_avg_CNH,_avg_WTI)
dist_HG = dtw.distance(_avg_CNH,_avg_HG)
dist_NG = dtw.distance(_avg_CNH,_avg_NG)

print(dist_GC)
print(dist_SI)
print(dist_WTI)
print(dist_HG)
print(dist_NG)
```
