```python
## CNH = 역외 위안화 환율
CNH = pd.read_csv('C:/Users/nueva/Desktop/p1/test/USD_CNH 과거 데이터.csv')

# 종가, 시가 분리 및 평균
CNH_Ed = CNH.iloc[:, 1]
CNH_St = CNH.iloc[:, 2]
avg_CNH = (CNH_Ed + CNH_St) /2 
avg_CNH = np.array(avg_CNH).reshape(-1,1)

# Fit_transform
scaler = StandardScaler()
_avg_CNH = scaler.fit_transform(avg_CNH)
```

```python
## DNX = 달러 인덱스 
DNX = pd.read_csv('C:/Users/nueva/Desktop/p1/test/달러 지수 과거 데이터.csv')

# DNX 종가, 시가 분리 및 평균
DNX_Ed = DNX.iloc[:, 1]
DNX_St = DNX.iloc[:, 2]
avg_DNX = (DNX_Ed + DNX_St) /2 
avg_DNX = np.array(avg_DNX).reshape(-1,1)

# Fit_transform
scaler = StandardScaler()
_avg_DNX = scaler.fit_transform(avg_DNX)
```

```python
## EUR = 유로 
EUR = pd.read_csv('C:/Users/nueva/Desktop/p1/test/EUR_USD 과거 데이터.csv')

# 종가, 시가 분리 및 평균
EUR_Ed = EUR.iloc[:, 1]
EUR_St = EUR.iloc[:, 2]
avg_EUR = (EUR_Ed + EUR_St) /2 
avg_EUR = np.array(avg_EUR).reshape(-1,1)

# Fit_transform
scaler = StandardScaler()
_avg_EUR = scaler.fit_transform(avg_EUR)
```

```python
## AUD = 호주 달러 
AUD = pd.read_csv('C:/Users/nueva/Desktop/p1/test/AUD_USD 과거 데이터.csv')

# 종가, 시가 분리 및 평균
AUD_Ed = AUD.iloc[:, 1]
AUD_St = AUD.iloc[:, 2]
avg_AUD = (AUD_Ed + AUD_St) /2 
avg_AUD = np.array(avg_AUD).reshape(-1,1)

# Fit_transform
scaler = StandardScaler()
_avg_AUD = scaler.fit_transform(avg_AUD)
```

```python
## JPY = 엔화 
JPY = pd.read_csv('C:/Users/nueva/Desktop/p1/test/USD_JPY 과거 데이터.csv')

# 종가, 시가 분리 및 평균
JPY_Ed = JPY.iloc[:, 1]
JPY_St = JPY.iloc[:, 2]
avg_JPY = (JPY_Ed + JPY_St) /2 
avg_JPY = np.array(avg_JPY).reshape(-1,1)

# Fit_transform
scaler = StandardScaler()
_avg_JPY = scaler.fit_transform(avg_JPY)
```

```python
# Display the distance
dist_DNX = dtw.distance(_avg_CNH,_avg_DNX)
dist_EUR = dtw.distance(_avg_CNH,_avg_EUR)
dist_AUD = dtw.distance(_avg_CNH,_avg_AUD)
dist_JPY = dtw.distance(_avg_CNH,_avg_JPY)

print(dist_DNX)
print(dist_EUR)
print(dist_AUD)
print(dist_JPY)
```
